```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yaml
from yaml.loader import FullLoader
```

```python
def get_CF_path(config_path):
    split = list(config_path.split('/'))
    relative_path = split[:-1]
    string_path = '/'.join(relative_path)

    with open(config_path) as f:
        config = yaml.load(f, Loader=FullLoader)

    weights = config['weights']
    epoch = weights.split('/')[-1].split('-')[1]
    CF_name =  f'epoch{epoch}_test_each_class_acc.csv'
    CF_path = string_path + '/' + CF_name

    CF = pd.read_csv(CF_path, header=None)
    current_col = CF.loc[0, :]
    current_col
    new_col = [str(round(float(i)*100, 1)) for i in current_col]
    CF.columns = [i+1 for i in range(len(current_col))]
    return CF
```

```python
config_path1 = 'pretrained/baseline/bone/work_dir/ntu_B/cview/ctrgcn/config.yaml'

CF1 = get_CF_path(config_path1)
# CF2 = get_CF_path(config_path2)
CF = CF1
confusion = CF.loc[1:, :]

plt.figure(figsize=(18, 12))
ax = sns.heatmap(confusion, cmap='Blues')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
config_path1 = 'pretrained/baseline/joint/work_dir/ntu_J/cview/ctrgcn/config.yaml'
config_path2 = 'pretrained/shift/joint/work_dir/ntu_J/cview/ctrgcn/config.yaml'

CF1 = get_CF_path(config_path1)
CF2 = get_CF_path(config_path2)
CF = CF2.sub(CF1)
confusion = CF.loc[1:, :]

plt.figure(figsize=(18, 12))
ax = sns.heatmap(confusion, cmap='PiYG')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```
